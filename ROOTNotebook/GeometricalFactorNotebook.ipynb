## Inizio analisi interattiva
A questo punto abbiamo eseguito la macro che permette di caricare i dati come desiderato.
Nella gDirectory abbiamo tutti gli oggetti generati dalla Macro. Si possono quindi eseguire grafici, fit, etc.

```c++
TCanvas *cAll = new TCanvas("","", 1000,600);

TColor *color_0 = new TColor(1756, 242./255., 53./255, 141./255);
TColor *color_1 = new TColor(1757, 4./255., 104./255, 191./255);
TColor *color_2 = new TColor(1758, 242./255., 152./255, 73./255);



hFinal2_0 -> SetLineColorAlpha(1756,1.);
hFinal2_0 -> SetMarkerStyle(20);
hFinal2_0 -> SetMarkerColorAlpha(1756,1.);


hFinal2_1 -> SetLineColorAlpha(1757,1.);
hFinal2_1 -> SetMarkerStyle(21);
hFinal2_1 -> SetMarkerColorAlpha(1757,1.);


hFinal2_2 -> SetLineColorAlpha(1758,1.);
hFinal2_2 -> SetMarkerStyle(22);
hFinal2_2 -> SetMarkerColorAlpha(1758,1.);






hFinal2_0 -> Draw("esame");
hFinal2_1 -> Draw("esame");
hFinal2_2 -> Draw("esame");

double Ymin = 0.001;
double Ymax = 0.6;

hFinal2_0 -> GetYaxis() -> SetRangeUser(Ymin, Ymax);
hFinal2_1 -> GetYaxis() -> SetRangeUser(Ymin, Ymax);
hFinal2_2 -> GetYaxis() -> SetRangeUser(Ymin, Ymax);

TLegend  *leg = new TLegend(0.1, 0.1, 0.3, 0.3);
leg -> AddEntry(hFinal2_0, "electrons", "lep");
leg -> AddEntry(hFinal2_1, "protons", "lep");
leg -> AddEntry(hFinal2_2, "alpha", "lep");

leg -> Draw();


gPad -> SetLogx();
gPad -> SetLogy();
gPad -> SetGrid();
cAll -> Draw();

cAll -> SaveAs("../docs/assets/GeometricalFactor/GeomFactor.C");
cAll -> SaveAs("../docs/assets/GeometricalFactor/GeomFactor.pdf");
cAll -> SaveAs("../docs/assets/GeometricalFactor/GeomFactor.png");
cAll -> SaveAs("../docs/assets/GeometricalFactor/GeomFactor.svg");
```

## Elettroni

```c++
TCanvas *c1 = new TCanvas();
hFinal_0 -> Draw();
gPad -> SetGrid();

TF1 *f1 = new TF1("f1", "[0] + [1]*(x - 3.)", 0.,12.);
hFinal_0 -> Fit("f1", "", "G", 1.8, 4.2);

c1 -> Draw();

cout << Form("p0 = (%g +/- %g)", f1 -> GetParameter(0), f1 -> GetParError(0)) << endl;
cout << Form("p1 = (%g +/- %g)", f1 -> GetParameter(1), f1 -> GetParError(1)) << endl;
cout << Form("Delta G = (%g +/- %g)", (4.2-1.8)*(f1->GetParameter(1)), (4.2-1.8)*(f1->GetParError(1))) << endl;
```

```c++
TCanvas *c1 = new TCanvas();
hFinal_1 -> Draw();
gPad -> SetGrid();

TF1 *f1 = new TF1("f1", "[0] + [1]*(x - 30.)", 0.,400.);
hFinal_1 -> Fit("f1", "", "G", 15, 50);

c1 -> Draw();

cout << Form("p0 = (%g +/- %g)", f1 -> GetParameter(0), f1 -> GetParError(0)) << endl;
cout << Form("p1 = (%g +/- %g)", f1 -> GetParameter(1), f1 -> GetParError(1)) << endl;
cout << Form("Delta G = (%g +/- %g)", (4.2-1.8)*(f1->GetParameter(1)), (4.2-1.8)*(f1->GetParError(1))) << endl;
```
